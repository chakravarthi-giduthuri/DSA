# BINARY SEARCH 

```python
def locate_card(cards,query):
    l,r = 0 ,len(cards)-1
    while l<=r:
        
        mid = (l+r)//2
        if cards[mid] == query:
            return mid
        elif cards[mid]<query:
            r = mid-1
        elif cards[mid]>query:
            l = mid+1
    return -1
```

```python
evaluate_test_cases(locate_card,tests)
```

```python
def test_location(cards,query,mid):
    if cards[mid] == query:
        if mid-1>=0 and cards[mid-1]==query:
            return 'left'
        else:
            return 'found'
    elif cards[mid]<query:
        return 'left'
    else:
        return 'right'
    
    
def locate_card(cards,query):
    lo,ho = 0,len(cards)-1
    while lo<=ho:
        mid = (lo+ho)//2
        result = test_location(cards,query,mid)
        if result == 'found':
            return mid
        elif result == 'left':
            ho = mid-1
        elif result == 'right':
            lo = mid+1
    return -1
```

```python
def binary_search(low,high,conditon):
    while low<=high:
        mid = (low+high)//2
        result = condition(mid)
        if result =='found':
            return mid
        elif result == 'left':
            high = mid-1
        elif result =='right':
            low = mid+1
    return -1

def first_position(cards,query):
    def condition(mid):
        if cards[mid] == query:
            if mid>0 and cards[mid-1]==query:
                return 'left'
            return 'found'
        elif cards[mid]<query:
            return 'right'
        else:
            return 'left'
    return binary_search(0,len(cards)-1,condtion)

def last_position(cards,query):
    def condition(mid):
        if cards[mid] == query:
            if mid<len(cards)-1 and cards[mid+1] == query:
                return 'right'
            return 'found'
        elif cards[mid]<query:
            return 'right'
        else:
            return 'left'
    return binary_search(0,len(cards)-1,condition)
    
def first_and_last_position(cards,query):
    return first_position(cards,query),last_position(cards,query)
    
        
    
```
